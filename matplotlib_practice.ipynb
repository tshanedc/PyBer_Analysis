```python
%matplotlib inline
```

```python
#Import Dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
#Change the line color to green.
ax.plot(x_axis, y_axis, color = "green")
#Change the marker to a diamond.
ax.plot(x_axis, y_axis, marker="D",color = "green")
#Change the line width to 2 points.
ax.plot(x_axis, y_axis, marker="D",color = "green", linewidth = 2)
#Add a legend for the city of Boston.
ax.plot(x_axis, y_axis, marker="D",color = "green", linewidth = 2, label = "Boston")
ax.legend()
#Add a title and axes labels.
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
#Add a y-axis limit.
ax.set_ylim(-10, 50)
#Add grid lines.
ax.grid()
```

```python
# Create a vertical bar chart
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.gca().invert_yaxis()
```

```python
# Skill Drill. 
fig, ax = plt.subplots()
#Change Color to Cyan

#Add a legend for the City of Chicago, label added before legend

#Add a title and axes labels
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
# Switch the axis so that Fare is on X-Axis
ax.barh(x_axis, y_axis, color = "cyan", label= "Chicago")
ax.legend()
#invert Y Axis
w = ax.invert_yaxis() 
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
